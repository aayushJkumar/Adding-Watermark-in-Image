```python
import numpy as np
import cv2 as cv
import os
import random
from PIL import Image
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
text=['SONY','REDMI','SAMSUNG','VIVO','OPPO','BLACKBERRY','NOKIA','PHILIPS','MOTOROLA','LAVA','APPLE','HUAWEI','PANASONIC','ONEPLUS']
```

```python
len(text)
```

```python
random.choice(text)
```

```python
images_tagged=os.listdir('/content/drive/MyDrive/IIT GUWAHATI ADDING WATERMARK TASK/Untagged Image')
img_tagged=[]
images_text=[]

for i in images_tagged :
  img=cv.imread('/content/drive/MyDrive/IIT GUWAHATI ADDING WATERMARK TASK/Untagged Image'+'/'+i)
  img=cv.resize(img,(900,700))
  img = cv.putText(img,random.choice(text),(10,500),cv.FONT_HERSHEY_SIMPLEX,2,(255,255,255),4,cv.LINE_AA)
  img_tagged.append(img)
  cv.imwrite(os.path.join('/content/drive/MyDrive/IIT GUWAHATI ADDING WATERMARK TASK/untagged final images',i),img)
 
  


```
